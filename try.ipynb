```python
import numpy as np
import time

M = np.deg2rad(7)
e = 0.999

f = lambda E : E - e * np.sin(E) - M
df = lambda E : 1 - e * np.cos(E)

# E0 = np.deg2rad(38.52700657)
B= 1.173439404
A = -0.584013113
C = 0.809460441
D = 0.077357763

a = (B * np.sin(M) + D * np.cos(M)) / ( (1 / e) - A * np.sin(M) - C * np.cos(M) )
E0 = M  + e * (np.sin( M + e * np.sin(M + a)))
E0
```

```python
from algo.newton import newtons_method

start = time.perf_counter()
newtons_method(f, df,  E0, 1e-5, 100)
(time.perf_counter() - start) * 1000_000
```

```python
from algo.fixed_point import fixed_point_iteration
g = lambda E : M + e * np.sin(E)
start = time.perf_counter()
fixed_point_iteration(f, g, E0, 1e-5, 100)
(time.perf_counter() - start) *  1000_000

```

```python
from algo.secant import secant_method

start = time.perf_counter()
secant_method(f,E0, E0 + .1,1e-5, 100)
(time.perf_counter() - start) *  1000_000

```

```python
from algo.haley import solve_halley

ddf = lambda E: e * np.sin(E)
start = time.perf_counter()
solve_halley(f, df, ddf, E0)
(time.perf_counter() - start) *  1000_000
```
